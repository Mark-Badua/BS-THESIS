```python
# freq = (('daily', 95,118), ('weekly',15,16), ('hourly', 4370, 4373))

def get_daily_plot(data, low, high): #nonzero frequencies [low, high)
    y = np.fft.rfft(data.iloc[:,0] - data.iloc[:,0].mean())
    z = y.copy()
    z[:low] = 0
    z[high:] = 0
    filtered = np.fft.irfft(z)
    return filtered

def ifft_df(data, tup): #tup = col_name, low, high
    new_data = pd.DataFrame()
    for col_name, low, high in tup:
        new_data[col_name] = get_daily_plot(data, low, high)+ data.iloc[:,0].mean()
    # new_data = new_data.iloc[:150720,:]
    new_data.index = data.index
    new_data['total'] = data.iloc[:,0]
    return new_data
```

```python
def get_daily_plot(data, low, high): #boundaries low is included and high is excluded
    length = get_next_power2(data.shape[0]) 
    y = np.fft.rfft(data.iloc[:,0] - data.iloc[:,0].mean(), n = length)
    z = y.copy()
    z[:low] = 0
    z[high:] = 0
    filtered = np.fft.irfft(z)
    return filtered
```
