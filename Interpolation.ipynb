```python
def nmse(y_true, y_pred):
    differences = (y_true - y_pred)/y_true                      
    differences_squared = differences ** 2                    
    mean_of_differences_squared = np.mean(differences_squared)         
    return mean_of_differences_squared
```

```python
path = r"/cxnfs/Mark/Data/cleaned_data/"
pm10 = pd.read_csv(path + 'pm10_cleaned.csv', index_col = 0, parse_dates = True)
pm25 = pd.read_csv(path + 'pm25_cleaned.csv', index_col = 0, parse_dates = True)
pm1 = pd.read_csv(path+'pm1_cleaned.csv', index_col = 0, parse_dates = True)
```

```python
df_trial = pd.DataFrame()
subsets_trial = prep_for_interp(pm10, ' PM10 (µg/m^3)')
df_trial['non missing '] = [subset.dropna().shape[0] for subset in prep_for_interp(pm10, ' PM10 (µg/m^3)')]
df_trial['missing'] = index_misisngval_pair(subsets_trial)
df_trial['max time difference'] = ind_maxgap_pair(subsets_trial)
df_trial['max time difference'] = df_trial['max time difference'].round(decimals = 2)
df_trial['date range'] = [subset.index.to_series().astype(str)[0] + ' to ' +  subset.index.to_series().astype(str)[-1] for subset in prep_for_interp(pm10, ' PM10 (µg/m^3)')]
df_trial['sample interval (days)'] = [(subset.index[-1] - subset.index[0])/np.timedelta64(1, 'D') for subset in subsets_trial]
df_trial.index.name = 'Segment'
df_trial['sample interval (days)'] = df_trial['sample interval (days)'].round(decimals = 2)
df_trial
```
