```python
#The normalized MSE allows you to compare the error for sets with vary different value ranges
def nmse(y_true, y_pred):
    differences = (y_true - y_pred)/y_true                      
    differences_squared = differences ** 2                    
    mean_of_differences_squared = np.mean(differences_squared)         
    return mean_of_differences_squared
```

```python
import time
def interp_tot_time(data, col, segment_list, interp_func, specify = None):
    subset_list = prep_for_interp(data, col)
    x = 0
    for ind, data in enumerate(subset_list):
        if not(ind in segment_list):
            continue
        start_time = time.perf_counter()
        interp(data,interp_func, specify = specify)
        end_time = time.perf_counter()
        x+= (end_time - start_time)
    return x
```
