```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'/home/jovyan/complete_data.csv', index_col = 0, parse_dates =  True)
#cleaned data
pm10 = pd.read_csv('pm10.csv', index_col = 0, parse_dates = True)
pm25 = pd.read_csv('pm25.csv', index_col = 0, parse_dates = True)
pm1 = pd.read_csv('pm1.csv', index_col = 0, parse_dates = True)
#storing PM data in a dataframe along with other variables
data = pd.DataFrame()
data.index = df.index
data['pm10'] = pm10.iloc[:,0]
data['pm25'] = pm25.iloc[:,0]
data['pm1'] = pm1.iloc[:,0]
data['windspeed'] = df[' Windspeed (m/s)']
data['wind_dir'] = df[' WindDirect (Degr)']
prepandemic_data = data['2020-02-01':'2020-03-09']
prepandemic_data = prepandemic_data.dropna(how = 'any', axis = 'rows')
#classes were suspended last March 10, 2020 to March 14, 2020
ecq_data = data['2020-03-17':'2020-05-15'] #start of lockdown until end of ECQ
ecq_data = ecq_data.dropna(how = 'any', axis = 'rows')
```

```python
dr = 0.25
dtheta = 15

radius_list = np.arange(0, 7+dr, dr)
theta_list = np.arange(0, 360+dtheta, dtheta)
prob_df = pd.DataFrame(index = theta_list, columns = radius_list)
size_df = pd.DataFrame(index = theta_list, columns = radius_list)

for current_rad, next_rad in zip(radius_list[:-1], radius_list[1:]):
    filtered_radius = prepandemic_data[(prepandemic_data.windspeed >= current_rad) & (prepandemic_data.windspeed < next_rad)]
    if filtered_radius.shape[0] == 0:
        continue
    for current_theta, next_theta in zip(theta_list[:-1], theta_list[1:]):
        filtered_rad_theta = filtered_radius[(filtered_radius.wind_dir >= current_theta) & (filtered_radius.wind_dir < next_theta)]
        if filtered_rad_theta.shape[0] == 0:
            continue
        prob_df.loc[current_theta, current_rad] = (filtered_rad_theta.pm25 >= data.pm25.quantile(0.75)).sum() / filtered_rad_theta.pm25.shape[0]
        size_df.loc[current_theta, current_rad] = filtered_rad_theta.pm25.shape[0]
        

A, R = np.meshgrid(theta_list*np.pi/180, radius_list)

# plot
fig, ax = plt.subplots(subplot_kw=dict(projection="polar"))

pc = ax.pcolormesh(A, R, prob_df.iloc[:-1,:-1].fillna(np.nan).T,  cmap = 'magma_r', vmin = 0, vmax = 1)
cbar = fig.colorbar(pc)
cbar.ax.set_ylabel('Prob ($PM_{2.5}~\mu g/m^3 \geq 16.6)$')
ax.set_rticks(np.arange(0,7,1))
ax.grid(True)
ax.set_rmax(7)
ax.set_theta_direction(-1)
ax.set_theta_zero_location('N')
plt.show()
```
