```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the power law function
def power_law(x, a, b):
    return a * (x ** (-b))

# Perform the curve fitting
params, cov = curve_fit(power_law, pm1_prepandemic_spec.freq[1:], pm1_prepandemic_spec.abs_fourier_coeff_square[1:])

# Retrieve the fitted parameters
a_fit, b_fit = params

# Generate fitted curve
x_fit = np.linspace(0.00001, 1, 100)
y_fit = power_law(x_fit, a_fit, b_fit)

# Plot the data and fitted curve
plt.loglog(pm1_prepandemic_spec.freq[1:], pm1_prepandemic_spec.abs_fourier_coeff_square[1:],'k.', label='Data')
plt.loglog(x_fit, y_fit, 'r-', label='Fitted curve')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# Print the fitted parameters
print("Fitted parameters:")
print("a =", a_fit)
print("b =", b_fit)
```
